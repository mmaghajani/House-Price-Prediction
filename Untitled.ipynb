```python


lam=0.1
for col in ('MiscVal', 'PoolArea', 'LotArea', 'LowQualFinSF', '3SsnPorch',
       'KitchenAbvGr', 'BsmtFinSF2', 'EnclosedPorch', 'ScreenPorch',
       'BsmtHalfBath', 'MasVnrArea', 'OpenPorchSF', 'WoodDeckSF',
       'LotFrontage', 'GrLivArea', 'BsmtFinSF1', 'BsmtUnfSF', 'Fireplaces',
       'HalfBath', 'TotalBsmtSF', 'OverallCond', 'YearBuilt',
       'GarageYrBlt'):
    train[col]=boxcox1p(train[col],lam)
    test[col]=boxcox1p(test[col],lam)

```

```python


train.drop('Id', axis=1, inplace=True)
test.drop('Id', axis=1, inplace=True)

```

```python
x=train.drop('SalePrice', axis=1)
y=train['SalePrice']
test=test.drop('SalePrice', axis=1)
```

```python


sc=RobustScaler()
x=sc.fit_transform(x)
test=sc.transform(test)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import make_scorer
```

```python
def root_mean_squared_log_error(y_test, y_pred):
    return mean_squared_log_error(y_test, y_pred)
```

```python
X_train, X_validation, y_train, y_validation = train_test_split(x, y ,   test_size = 0.1, random_state = 26)
```

```python
from sklearn.model_selection import GridSearchCV
import pprint
```

```python
# Number of trees in random forest
# Number of features to consider at every split
alpha = []
for i in range(1, 1000, 100):
    alpha.append(i/1000)
    
tol = []
for i in range(1, 1000, 50 ):
    tol.append(i/10000)

selection = ['cyclic', 'random']
# Create the random grid
exaust_grid = {
               'alpha' : alpha,
                'tol' : tol,
                'selection': selection}
exaust_grid
```

```python
RMSLE = make_scorer(root_mean_squared_log_error)
regressor = Lasso(random_state=1) 
rf_random = GridSearchCV(scoring = RMSLE, estimator = regressor, param_grid = exaust_grid, cv = 5, verbose=2, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```
